```python
top_3_Cuisines =  df['Cuisines'].str.split(',').explode().value_counts().head(3)
top_3_Cuisines
```

```python
north_indian = df['Cuisines'].str.contains('North Indian').mean() * 100
chinese_cuisines = df['Cuisines'].str.contains('Chinese', case= False, na=False).mean() * 100
Fast_Food = df['Cuisines'].str.contains('Fast Food').mean() * 100
print(north_indian)
print(chinese_cuisines)
print(Fast_Food)
```

```python
names = ['north_indian','chinese_cuisines','Fast_Food']
values= [north_indian,chinese_cuisines,Fast_Food]
plt.bar(names, values, color = 'Blue')
plt.title('The percentage of restaurants that serve each of the top cuisines')
plt.ylabel('percentage (%)')
plt.show()
```
