```python
df['artist_genres'] = df['artist_genres'].fillna('')
df['artist_genres'] 
```

```python
for x in df.index:
    if df.loc[x, "artist_genres"] == '':
        df.drop(x,inplace=True)
df        
```

```python
df = df.drop(columns = ['track_id','track_number','explicit','album_id'])
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['artist_name'].nunique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.hist(figsize=(14,10), bins=30)
plt.suptitle('Numerical Feature Distributions')
plt.show()

if 'popularity' in df.columns:
    sns.histplot(df['popularity'], kde=True)
    plt.title('Popularity Distribution')
    plt.show()
```

```python
if 'artist_name' in df.columns:
    top_artist = df['artist_name'].value_counts().head(10)
    top_artist.plot(kind = 'bar', color = 'red')
    plt.title("TOP 10 ARTIST BY TRACK ENTRIES")
    plt.ylabel("Track Count")
    plt.xticks(rotation = 45)
    plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
top_tracks = df.nlargest(10, 'track_popularity')
sns.barplot(x='track_popularity', y='track_name', data=top_tracks)
plt.xlabel("Popularity")
plt.ylabel("Track Name")
plt.title("Top 10 Most Popular Tracks")

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='album_type', data=df, palette='bright')
plt.title("Number of Tracks per Album Type")
plt.xlabel("Album Type")
plt.ylabel("Count")
plt.show()
```
